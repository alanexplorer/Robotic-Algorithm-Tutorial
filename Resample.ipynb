```python
import numpy as np
from math import pi
import random
from copy import deepcopy
```

```python
M = 100
```

```python
class Particle():

    def __init__(self, M):

        self.weight = 1.0/M

    def get_weight(self):
        return self.weight
    
    def set_weight(self, w):
        self.weight *= w
        
    def norm_weight(self, n):
        self.weight = n
```

```python
particles = [Particle(M) for i in range(M)]
```

```python
global sum_weight
sum_weight = 0
for i in range(M):
    w = random.uniform(0, 1)
    particles[i].set_weight(w)
    sum_weight += particles[i].get_weight()
```

```python
for i in range(M):
    w = particles[i].get_weight()
    particles[i].norm_weight(w/sum_weight)
```

```python
for i in range(M):
    print(particles[i].get_weight())
```

```python
def resample_particles(particles):
    new_particles = []
    weight = [p.get_weight() for p in particles]
    index = int(random.random() * M)
    beta = 0.0
    mw = max(weight)
    for i in range(M):
        beta += random.random() * 2.0 * mw
        while beta > weight[index]:
            beta -= weight[index]
            index = (index + 1) % M
        new_particle = deepcopy(particles[index])
        new_particle.weight = 1
        new_particles.append(new_particle)
    return new_particles
```

```python
particles = resample_particles(particles)
for i in range(M):
    print(particles[i].get_weight())
```
